```python
from datasets import load_dataset

data = load_dataset("jamescalam/ai-arxiv2", split="train")
data
```

```python
content = data[3]["content"]
print(content[:1000])
```

```python
content = content[:20_000]
```

```python
!pip install transformers
import os
from getpass import getpass
# from semantic_router.encoders import OpenAIEncoder
from semantic_router.encoders import HuggingFaceEncoder


# os.environ["OPENAI_API_KEY"] = os.getenv("OPENAI_API_KEY") or getpass(
#     "OpenAI API key: "
# )

# encoder = OpenAIEncoder(name="text-embedding-3-small")
encoder = HuggingFaceEncoder()

```

**Statistical Chunking:**

Fast and cost effective. it uses a varying similarity threshold to identify more dynamic and local similarity splits. It offers a good balance between accuracy and efficiency. Only works on text.

```python
from semantic_chunkers import ConsecutiveChunker

chunker = ConsecutiveChunker(encoder=encoder, score_threshold=0.3)
chunks = chunker(docs=[content])
```

```python
chunker.print(chunks[0])
```

**Cumulative Chunking:**

Cumulative chunking is a more compute intensive process, but can often provide more stable results as it is more noise resistant.


```python
from semantic_chunkers import CumulativeChunker

chunker = CumulativeChunker(encoder=encoder, score_threshold=0.3)
chunks = chunker(docs=[content])
```

```python
chunker.print(chunks[0])
```

**Consecutive Chunking:**

It is multi modal chunker.


```python
from semantic_chunkers import ConsecutiveChunker

chunker = ConsecutiveChunker(encoder=encoder, score_threshold=0.3)
chunks = chunker(docs=[content])
```

```python
chunker.print(chunks[0])
```
